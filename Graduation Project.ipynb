# What is the sales trend over time (monthly, yearly)?

```python
# Convert 'Order Date' to datetime format
df['Order Date'] = pd.to_datetime(df['Order Date'], format='%d/%m/%Y')

# Monthly sales trend
monthly_sales = df.groupby(pd.Grouper(key='Order Date', freq='M'))[['Sales']].sum()

# Yearly sales trend
yearly_sales = df.groupby(pd.Grouper(key='Order Date', freq='Y'))[['Sales']].sum()

# Plotting monthly sales
plt.figure(figsize=(14, 6))

# Monthly Sales Plot
plt.subplot(1, 2, 1)
monthly_sales.plot(kind='line', marker='o', ax=plt.gca())
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.grid()

# Yearly Sales Plot
plt.subplot(1, 2, 2)
yearly_sales.plot(kind='bar', ax=plt.gca())
plt.title('Yearly Sales Trend')
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.xticks(rotation=0)
plt.grid()

plt.tight_layout()
plt.show()
```

# Calculate the Shipping Duration

```python
# Specify the date format for 'Order Date' and 'Ship Date'
df['Order Date'] = pd.to_datetime(df['Order Date'], errors='coerce')
df['Ship Date'] = pd.to_datetime(df['Ship Date'], format="%d/%m/%Y", errors='coerce')

# Check if there are any NaT values (invalid dates) in the 'Order Date' or 'Ship Date' columns
print("Missing 'Order Date':", df['Order Date'].isna().sum())
print("Missing 'Ship Date':", df['Ship Date'].isna().sum())
```

```python
# Calculate the shipping duration (in days)
df['Shipping Duration'] = (df['Ship Date'] - df['Order Date']).dt.days

# Display the first few rows to verify
print(df[['Order Date', 'Ship Date', 'Shipping Duration']].head())
```

```python
# Plot the distribution of shipping duration
plt.figure(figsize=(10, 6))
plt.hist(df['Shipping Duration'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Shipping Duration', fontsize=16)
plt.xlabel('Shipping Duration (Days)', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.grid(True)
plt.tight_layout()
plt.show()
```

The graph shows a distribution of shipping times from 0 to 7 days, with the majority of orders being shipped in 4 or 5 days.

# Is shipping duration affect my sales?

```python
# Calculate the total sales grouped by shipping duration
sales_duration = df.groupby('Shipping Duration')['Sales'].sum().reset_index()

# Calculate the correlation
correlation = sales_duration['Shipping Duration'].corr(sales_duration['Sales'])
print(f'Correlation between Shipping Duration and Sales: {correlation}')
```

Weak Positive Correlation:
The correlation coefficient is close to 0 but slightly positive.
This indicates that there is a weak tendency for sales to increase as
shipping duration increases. However, the relationship is not strong enough
to suggest that shipping duration has a significant impact on sales.

# 4. Analyzing by Date (Seasonal Patterns, Holidays)

```python
# Extract month and analyze if delays are seasonal
df['Order Month'] = df['Order Date'].dt.month
monthly_delay = df.groupby('Order Month')['Shipping Duration'].mean()

# Plot monthly shipping duration
sns.lineplot(x=monthly_delay.index, y=monthly_delay.values)
plt.title("Average Shipping Duration by Month")
plt.ylabel('Average Shipping Duration (Days)')
plt.xlabel('Month')
plt.show()
```

The graph shows the Average Shipping Duration by Month. Here's what I observe:

January and April: These months show higher shipping durations (around 4.2 days). This could be due to post-holiday season delays in January and possibly logistical adjustments in April (which may vary by company or regional conditions).

February: There's a significant drop in the average shipping duration to around 3.7 days, indicating faster processing during this period. It could be a quieter time of year in terms of orders, leading to quicker shipments.

Summer months (June-August): There's a steady increase, with shipping times rising slightly during these months, potentially reflecting increased demand or vacation schedules affecting the logistics process.

End of the Year (November-December): Theres a noticeable rise again during November and December, which could be linked to the holiday shopping season, where increased demand and potential logistical constraints lead to longer shipping times.

Insights:
Holidays and busy seasons (like November-December) seem to affect shipping durations, leading to delays.
January delays might be residual from the holiday rush, while the April spike might warrant further investigation (e.g., specific events or issues during that time).
February and other quieter months tend to have shorter shipping durations.
